```python
import re

def remove_spaces(s:str):
    """
    Arg: s (str): input string with possible excess spaces
    Ret: s (str): string free from excess spaces
    """
    s = " ".join(s.split())
    if s[0]==' ':
        s = re.sub(r"^\s+", "", s)
    if s[-1]==' ':
        re.sub(r"\s+$", "", s)
    return s
```

```python
characters_mbti_df['char'] = characters_mbti_df['char'].str.lower()
characters_mbti_df['movie'] = characters_mbti_df['movie'].str.lower()

with_char_mbti = response_details_merge.copy()

def get_char_mbti(char:str, movie:str):
    """"
    Args: 
        char (str): movie character
        movie (str): movie title
    Returns:
        str: mbti of the character in the movie
    """
    # Get boolean based on string presence condition
    bool_series_charmbti = characters_mbti_df['char'].str.contains(char) & characters_mbti_df['movie'].str.contains(movie)

    if bool_series_charmbti.any():
        index = bool_series_charmbti.idxmax()
        return characters_mbti_df.loc[index, 'mbti']  
    else:
        return None
    
# Iterate over all the entries of chars and movies
for i in range(len(with_char_mbti)):
    char = remove_spaces(with_char_mbti.loc[i, 'character'].lower())
    movie = remove_spaces(with_char_mbti.loc[i, 'movie'].lower())

    # Directly add the mbti to the 'char_mbti' column
    with_char_mbti.loc[i, 'char_mbti'] = get_char_mbti(char, movie)
```
