```python
response_details_merge = pd.read_csv('../data/Data - respondents details merge.csv')
response_details_merge.drop(columns='Unnamed: 0', inplace=True)
response_details_merge
```

```python
try_df = response_details_merge.head(20)
try_df = try_df.drop(['mbti_test', 'movie'], axis=1)
try_df
```

```python
characters_mbti_df = pd.read_csv('../data/Data - Movie Characters MBTI clean.csv')
# survey_responses_df.drop(columns='Unnamed: 0', inplace=True)
characters_mbti_df
```

```python
recorded_chars = []
recorded_movies = []
missing_chars = []
missing_movies = []

# Define the column headers
columns = ['char', 'movie', 'mbti', 'votes']

# Create an empty DataFrame with only headers
new_df = pd.DataFrame(columns=columns)

def char_mbti_search(char_find, movie_find):
    bool_series = characters_mbti_df['char'].str.contains(char_find) & characters_mbti_df['movie'].str.contains(movie_find)
    if bool_series.any():
        index = bool_series.idxmax()  # Or use bool_series[bool_series].index for all matches
        print(index)
        recorded_chars.append(char_find)
        recorded_movies.append(movie_find)
        new_row = characters_mbti_df.loc[index]  
        return new_row
    else:
        # print(f"No match found for character '{char_find}' in movie '{movie_find}'")
        missing_chars.append(char_find)
        missing_movies.append(movie_find)
        return None


# char_list = list(try_df['character'])
# movie_list = list(try_df['movie'])
# print(char_list)

# char_list = 'Celine'
# movie_list = 'Before'

# char_mbti_search(char_list, movie_list)
# char_mbti_search('Tony Stark', 'MCU')

for i in range(len(response_details_merge)):
    char_i = str(response_details_merge['character'][i])
    movie_i = str(response_details_merge['movie'][i])
    # print(type(char_i), movie_i)
    new_row = char_mbti_search(char_i, movie_i)
    try:
        new_df = pd.concat([new_df, new_row.to_frame().T], ignore_index=True)
    except:
        pass
    # print(new_row)
```
