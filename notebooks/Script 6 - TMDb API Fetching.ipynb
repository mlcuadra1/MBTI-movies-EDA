```python
characters_mbti_df = pd.read_csv('../data/Data 8 - Franchise-expanded Movies from Data 2.csv', encoding='unicode_escape')
characters_mbti_df
```

```python
recorded_chars = []
recorded_movies = []
missing_chars = []
missing_movies = []
characters_mbti_df['char'] = characters_mbti_df['char'].str.lower()
characters_mbti_df['movie'] = characters_mbti_df['movie'].str.lower()

def char_mbti_search(char_find, movie_find):
    bool_series = characters_mbti_df['char'].str.contains(char_find) & characters_mbti_df['movie'].str.contains(movie_find)
    if bool_series.any():
        index = bool_series.idxmax()  # Or use bool_series[bool_series].index for all matches
        # print(index)
        recorded_chars.append(characters_mbti_df.loc[index])
        recorded_movies.append(movie_find)
        return characters_mbti_df.loc[index]  
    else:
        # print(f"No match found for character '{char_find}' in movie '{movie_find}'")
        missing_chars.append(char_find)
        missing_movies.append(movie_find)
        return None


# char_list = list(try_df['character'])
# movie_list = list(try_df['movie'])
# print(char_list)

# char_list = 'Tony'
# movie_list = 'Iron Man'

# print(char_mbti_search('Gazelle', 'Zootopia'))

import re
def remove_spaces(s):
    s = " ".join(s.split())
    if s[0]==' ':
        s = re.sub(r"^\s+", "", s)
    if s[-1]==' ':
        re.sub(r"\s+$", "", s)
    return s

for i in range(len(response_details_merge)):
    char_i = remove_spaces(str(response_details_merge['character'][i]).lower())
    movie_i = remove_spaces(str(response_details_merge['movie'][i]).lower())

    try:
        movie_split = movie_i.split(' (', maxsplit=1)
        movie_i = movie_split[0]
    except:
        pass
    
    char_mbti_search(char_i, movie_i)
```
