```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Import selenium necessities
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Set up the Chrome WebDriver
options = webdriver.ChromeOptions()
prefs = {"profile.managed_default_content_settings.images": 2}
options.add_experimental_option("prefs", prefs)
# options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--disable-blink-features=AutomationControlled')
```

```python
driver = webdriver.Chrome(options=options)
URL = "https://www.personality-database.com/profile?pid=2&cid=27&group=*%20Movies%20%5BMCU%5D"
driver.get(URL)
# time.sleep(20)
```

```python
char_full_list = []
movie_full_list = []
mbti_full_list = []
vote_full_list = []
```

```python
char_list = []
movie_list = []
mbti_list = []
vote_list = []
# Define a function to extract data from current page
def extract_page_data(driver):

    # Locate all elements on the current page
    page_elements = driver.find_elements(By.XPATH, '//*[@id="root"]/div/section/main/div[1]/div[2]/div/div[4]')

    char_element = page_elements[0].find_elements(By.CLASS_NAME, 'info-name')
    movie_element = page_elements[0].find_elements(By.CLASS_NAME, 'info-subcategory')
    mbti_element = page_elements[0].find_elements(By.CLASS_NAME, 'personality')
    vote_element = page_elements[0].find_elements(By.CLASS_NAME, 'vote-count')

    for i in range(69):
        # print(i)
        char_list.append(char_element[i].get_attribute('innerHTML'))
        # Extract 'label' from 'selenium.webdriver.remote.webelement.WebElement'
        movie_list.append(movie_element[i].text)
        mbti_list.append(mbti_element[i].get_attribute('innerHTML'))
        try:
            vote_list.append(vote_element[i].text)
        except:
            vote_list.append('0')
    return char_list, movie_list, mbti_list, vote_list

char, movie, mbti, votes = extract_page_data(driver)
df = pd.DataFrame({'char': char, 'movie': movie, 'mbti': mbti, 'votes': votes})
df
```

```python
char_full_list.extend(char_list)
movie_full_list.extend(movie_list)
mbti_full_list.extend(mbti_list)
vote_full_list.extend(vote_list)

df1 = pd.DataFrame({'char': char_full_list, 'movie': movie_full_list, 'mbti': mbti_full_list, 'votes': vote_full_list})
df1
```
